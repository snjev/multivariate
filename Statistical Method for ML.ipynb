```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x_axis = np.arange(-3,3,0.001)
y_axis = norm.pdf(x_axis,0,1)
plt.plot(x_axis,y_axis)
plt.show()
```

```python
np.random.RandomState(seed=1)
data = 5*np.random.randn(10000)+50
plt.hist(data)
plt.show()
```

```python
np.random.RandomState(seed=1)
data = 5*np.random.randn(10000)+50
plt.hist(data,bins=100)
plt.show()
```

```python
np.random.RandomState(seed=1)
data = 5*np.random.randn(10000)+50
result = np.mean(data)
print('Mean: %.3f'%result)
result_median = np.median(data)
print('Median: %.3f'%result_median)
res_var = np.var(data)
print('Variance: %.3f'%res_var)
res_std = np.std(data)
print('Variance: %.3f'%res_std)
```
