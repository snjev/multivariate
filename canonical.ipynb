```python
import numpy as np
import pandas as pd
import seaborn as sb
dataset = pd.read_csv('Facebook_metrics/dataset_Facebook.csv',delimiter=';')
#del dataset['Page total likes']
del dataset['Type']
del dataset['Paid']
#del dataset['Total Interactions']
dataset = dataset.dropna(how='any',axis=0)
mat_1 = dataset.iloc[:,[0,-1]].T
mat_2  =dataset.iloc[:,1:-2].T
dataset = dataset.T

cov_mat = np.cov(dataset)
s_xx = cov_mat[[0,-1],0:2]
s_xy = cov_mat[[0,-1],2:-1]
s_yx = s_xy.T
s_yy = cov_mat[2:-1,2:-1]

w_xx, v_xx = np.linalg.eig(s_xx)
w_xx_inv = w_xx**-0.5
w_xx_inv_p = np.diag(w_xx_inv).reshape(2,2)
e_xx = np.matmul(np.matmul(v_xx,w_xx_inv_p),v_xx.T)

w_yy, v_yy = np.linalg.eig(s_yy)
w_yy_inv = w_yy**-1
w_yy_inv_p = np.diag(w_yy_inv).reshape(14,14)
e_yy = np.matmul(np.matmul(v_yy,w_yy_inv_p),v_yy.T)

m1=np.matmul(e_xx,s_xy)
m2=np.matmul(m1,e_yy)
m3=np.matmul(m2,s_yx)
m4=np.matmul(m3,e_xx)

S = m4

w_S,v_S = np.linalg.eig(S)
w_S_1 = np.reshape(w_S, (1, 2))
#eigen values of S i.e w_S mul XX^1/2 mul X
w_xx, v_xx = np.linalg.eig(s_xx)
w_xx_inv_1 = w_xx**0.5
w_xx_inv_p1 = np.diag(w_xx_inv_1).reshape(2,2)
e_xx_1 = np.matmul(np.matmul(v_xx,w_xx_inv_p1),v_xx.T)

final_1 = np.matmul(np.matmul(w_S_1,e_xx_1),mat_1)
final_1 = final_1.T

#b1
w_yy, v_yy = np.linalg.eig(s_yy)
w_yy_inv = w_yy**-0.5
w_yy_inv_p = np.diag(w_yy_inv).reshape(14,14)
e_yy = np.matmul(np.matmul(v_yy,w_yy_inv_p),v_yy.T)

f_k = np.matmul(np.matmul(np.matmul(w_S_1,e_xx),s_xy),s_yy)
v_1 = np.matmul(np.matmul(f_k,e_yy),mat_2) 
v_1=v_1.T
```
